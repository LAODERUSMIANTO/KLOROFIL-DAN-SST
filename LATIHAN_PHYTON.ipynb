```python
!pip install xarray cartopy matplotlib
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import numpy as np

file_path = '/content/drive/MyDrive/phyton/data klorofil.nc'
ds = xr.open_dataset(file_path)

# Tampilkan struktur dataset
variable = ds['chlor_a']  # Ganti dengan variabel di dataset Anda
lats = ds['lat']
lons = ds['lon']
plt.figure(figsize=(12, 6))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_extent([95, 141, -11, 6], crs=ccrs.PlateCarree())  # Batas peta (contoh)

# Tentukan levels untuk colorbar
levels = np.arange(0, 11, 1)  # Rentang nilai dari 0 hingga 20 dengan interval 1

# Tambahkan data dengan pengaturan colorbar
im = variable.plot(ax=ax,
                  cmap='jet',
                  transform=ccrs.PlateCarree(),
                  levels=levels,  # Menggunakan levels yang telah ditentukan
                  cbar_kwargs={'label': 'Konsentrasi Klorofil (mg/m^3)',
                               'orientation': 'horizontal',
                               'pad': 0.05,
                               'shrink': 0.8,
                               'extend': 'both'})  # Menambahkan panah di ujung colorbar jika data di luar rentang


# Tambahkan fitur peta
ax.coastlines()
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.LAND, edgecolor='green')
ax.add_feature(cfeature.OCEAN)

# Tambahkan judul
plt.title('Peta Konsentrasi Klorofil')
plt.show()
```

# **SST**

```python
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import numpy as np
import cartopy.feature as cfeature

# Path file NetCDF
file_path = '/content/drive/MyDrive/phyton/sst.nc'
ds = xr.open_dataset(file_path)

# Menampilkan struktur dataset
print(ds)

# Ambil variabel SST dan koordinat
sst = ds['sst']  # Pastikan ini adalah variabel yang benar
lats = ds['lat']
lons = ds['lon']

# Ambil data SST (langsung tanpa memilih dimensi waktu)
sst_values = sst.values  # Seluruh data SST untuk lat dan lon yang ada

# Menambahkan noise untuk efek tekstur kasar (misalnya, dengan distribusi acak)
noise = np.random.normal(scale=0.5, size=sst_values.shape)  # Menambahkan noise acak
sst_noisy = sst_values + noise  # Kombinasikan data SST dengan noise

# Buat meshgrid untuk koordinat
lon_mesh, lat_mesh = np.meshgrid(lons, lats)

# Plot peta dengan koordinat dan data SST
plt.figure(figsize=(12, 6))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_extent([119, 125, -11, -8], crs=ccrs.PlateCarree())  # Batas Laut Sawu

# Gunakan pcolormesh untuk plot data SST yang telah ditambahkan noise
im = ax.pcolormesh(lon_mesh, lat_mesh, sst_noisy, cmap='jet', transform=ccrs.PlateCarree(), shading='auto')

# Menambahkan fitur-fitur peta
ax.coastlines()  # Menambahkan garis pantai
ax.add_feature(cfeature.BORDERS, linestyle=':')  # Menambahkan batas negara dengan garis putus-putus
ax.add_feature(cfeature.LAND, edgecolor='black')  # Menambahkan daratan dengan warna hijau
ax.add_feature(cfeature.OCEAN)  # Menambahkan lautan (dengan warna biru muda secara default)

# Menambahkan colorbar
plt.colorbar(im, ax=ax, orientation='vertical', label='Sea Surface Temperature (°C)')

# Menampilkan peta
plt.show()
```

```python
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import numpy as np
import cartopy.feature as cfeature

# Path file NetCDF
file_path = '/content/drive/MyDrive/phyton/sst.nc'
ds = xr.open_dataset(file_path)

# Menampilkan struktur dataset
print(ds)

# Ambil variabel SST dan koordinat
sst = ds['sst']  # Pastikan ini adalah variabel yang benar
lats = ds['lat']
lons = ds['lon']

# Ambil data SST (langsung tanpa memilih dimensi waktu)
sst_values = sst.values  # Seluruh data SST untuk lat dan lon yang ada

# Menambahkan noise untuk efek tekstur kasar (misalnya, dengan distribusi acak)
noise = np.random.normal(scale=0.5, size=sst_values.shape)  # Menambahkan noise acak
sst_noisy = sst_values + noise  # Kombinasikan data SST dengan noise

# Buat meshgrid untuk koordinat
lon_mesh, lat_mesh = np.meshgrid(lons, lats)

# Plot peta dengan koordinat dan data SST
plt.figure(figsize=(12, 6))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_extent([119, 125, -11, -8], crs=ccrs.PlateCarree())  # Batas Laut Sawu

# Gunakan pcolormesh untuk plot data SST yang telah ditambahkan noise
im = ax.pcolormesh(lon_mesh, lat_mesh, sst_noisy, cmap='jet', transform=ccrs.PlateCarree(), shading='auto')

# Menambahkan fitur-fitur peta
ax.coastlines()  # Menambahkan garis pantai
ax.add_feature(cfeature.BORDERS, linestyle=':')  # Menambahkan batas negara dengan garis putus-putus
ax.add_feature(cfeature.LAND, edgecolor='black')  # Menambahkan daratan dengan warna hijau
ax.add_feature(cfeature.OCEAN)  # Menambahkan lautan (dengan warna biru muda secara default)

# Menambahkan grid pada peta
gridlines = ax.gridlines(draw_labels=True, linestyle="--", color="gray", alpha=0.5)
gridlines.top_labels = False  # Menonaktifkan label di bagian atas
gridlines.right_labels = False  # Menonaktifkan label di sisi kanan
gridlines.xlabel_style = {'size': 10, 'color': 'black'}  # Gaya label bujur
gridlines.ylabel_style = {'size': 10, 'color': 'black'}  # Gaya label lintang

# Menambahkan colorbar di bawah peta
cbar = plt.colorbar(im, ax=ax, orientation='horizontal', pad=0.05, fraction=0.046, shrink=0.8)
cbar.set_label('Sea Surface Temperature (°C)', fontsize=12)

# Menampilkan peta
plt.tight_layout()  # Menyesuaikan layout agar tidak terpotong
plt.show()
```
